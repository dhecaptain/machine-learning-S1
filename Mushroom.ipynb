**A small description of the columns**
- **class**: Edible (e) or poisonous (p) 
- **cap_shape**: Shape of the cap (bell=b, conical=c, convex=x, flat=f, knobbed=k, sunken=s)
- **cap_surface**: Surface of the cap (fibrous=f, grooves=g, scaly=y, smooth=s)
- **cap_color**: Color of the cap (brown=n, buff=b, cinnamon=c, gray=g, green=r, pink=p, purple=u, red=e, white=w, yellow=y)
- **bruises**: Whether there are bruises (bruises=t, no=f)
- **odor**: Odor (almond=a, anise=l, creosote=c, fishy=y, foul=f, musty=m, none=n, pungent=p, spicy=s)
- **gill_attachment**: Gill attachment (attached=a, descending=d, free=f, notched=n)
- **gill_spacing**: Gill spacing (close=c, crowded=w, distant=d)
- **gill_size**: Gill size (broad=b, narrow=n)
- **gill_color**: Gill color (black=k, brown=n, buff=b, chocolate=h, gray=g, green=r, orange=o, pink=p, purple=u, red=e, white=w, yellow=y)
- **stalk_shape**: Stalk shape (enlarging=e, tapering=t)
- **stalk_root**: Stalk root (bulbous=b, club=c, cup=u, equal=e, rhizomorphs=z, rooted=r, missing=?)
- **stalk_surface_above_ring**: Stalk surface above the ring (fibrous=f, scaly=y, silky=k, smooth=s)
- **stalk_surface_below_ring**: Stalk surface below the ring (fibrous=f, scaly=y, silky=k, smooth=s)
- **stalk_color_above_ring**: Stalk color above the ring (brown=n, buff=b, cinnamon=c, gray=g, orange=o, pink=p, red=e, white=w, yellow=y)
- **stalk_color_below_ring**: Stalk color below the ring (brown=n, buff=b, cinnamon=c, gray=g, orange=o, pink=p, red=e, white=w, yellow=y)
- **veil_type**: Veil type (partial=p, universal=u)
- **veil_color**: Veil color (brown=n, orange=o, white=w, yellow=y)
- **ring_number**: Number of rings (none=n, one=o, two=t)
- **ring_type**: Type of ring (cobwebby=c, evanescent=e, flaring=f, large=l, none=n, pendant=p, sheathing=s, zone=z)
- **spore_print_color**: Spore print color (black=k, brown=n, buff=b, chocolate=h, green=r, orange=o, purple=u, white=w, yellow=y)
- **population**: Population (abundant=a, clustered=c, numerous=n, scattered=s, several=v, solitary=y)
- **habitat**: Habitat (grasses=g, leaves=l, meadows=m, paths=p, urban=u, waste=w, woods=d)

**Data Exploration**

UNIVARIATE ANALYSIS

```python
for col in data.columns:
    print(data[col].value_counts())
```

```python
import math

# Set up the matplotlib figure
num_columns = len(data.columns)
num_rows = math.ceil(num_columns / 2)

fig, axes = plt.subplots(nrows=num_rows, ncols=2, figsize=(14, num_rows * 4))
fig.tight_layout(pad=5.0)

# List of columns to visualize
columns = data.columns

# Loop through columns and plot
for i, column in enumerate(columns):
    sns.countplot(data=data, x=column, ax=axes[i//2, i%2])
    axes[i//2, i%2].set_title(f'Distribution of {column}')
    axes[i//2, i%2].set_xlabel('')
    axes[i//2, i%2].set_ylabel('')

# Remove empty subplots
for j in range(i + 1, num_rows * 2):
    fig.delaxes(axes.flatten()[j])

plt.show()
```

**BIVARIATE ANALYSIS**

```python
# Analyze the relationship between pairs of variable
# For example the relationship between class and other features.

#Relationship between 'class' and 'cap-shape'
sns.countplot(x='cap-shape',hue='class',data=data)
plt.show()
```

```python
# Relationship between class and cap-surface
sns.countplot(x='cap-surface',hue='class',data=data)
plt.show()
```

```python
# Ensure 'class' and other features are treated as categorical
for column in data.columns:
    data[column] = data[column].astype('category')

# Set up the matplotlib figure
num_columns = len(data.columns) - 1  # Exclude 'class' column
num_rows = math.ceil(num_columns / 2)

fig, axes = plt.subplots(nrows=num_rows, ncols=2, figsize=(14, 28))
fig.tight_layout(pad=5.0)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop through columns and plot
for i, column in enumerate(data.columns[1:]):  # Skip 'class' column itself
    sns.countplot(data=data, x=column, hue='class', ax=axes[i])
    axes[i].set_title(f'{column} by Class')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.show()
```

**MULTIVARIATE ANALYSIS**

```python
# Encode categorical variables as numbers for correlation analysis
from sklearn.preprocessing import LabelEncoder

# Create a new DataFrame with encoded values
encoded_df = data.apply(LabelEncoder().fit_transform)

# Calculate the correlation matrix
corr_matrix = encoded_df.corr()

# Plot the heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix of Encoded Features')
plt.show()
```
