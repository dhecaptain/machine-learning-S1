## Hyperparameter Tuning:
    1. GridSearchCV
    2. RandomizedSearchCV

```python
# importing necessary modules
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

the dataset will be of breast cancer

```python
# loading the dataset
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
# loading the data into a dataframe
data = pd.DataFrame(breast_cancer.data,columns=breast_cancer.feature_names)
```

```python
data.head()
```

```python
# adding the label 'target' column into the data
data['target'] = breast_cancer.target
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
# check for missing values
data.isnull().sum()
```

```python
# checking for the distribution of the target variable
data["target"].value_counts()
```

```python
# separating the data 
X =  data.drop("target",axis=1)
y = data["target"]
```

```python
X.head()
```

```python
y.head()
```

```python
X = np.asarray(X)
y = np.asarray(y)
X
```

## GridSearchCV

GridSearchCV is used for determining the best parameter for the model

```python
# loading the SVC
model = SVC()
```

```python
# hyperparameters

parameters = {
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
# grid search 
classifier = GridSearchCV(model,parameters,cv=5)
```

```python
# fitting the data to our model
classifier.fit(X,y)
```

```python
classifier.cv_results_
```

```python
# Best Parameter 

best_parameters = classifier.best_params_
best_parameters
```

```python
# Determining the highest accuracy
highest_accuracy = classifier.best_score_
highest_accuracy
```

```python
# load the results to pandas dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head(8)
```

```python
grid_search_result =result[["param_C","param_kernel","mean_test_score"]]
```

```python
grid_search_result
```

Highest Accuracy = 95.2%

Best Parameters = {'C':10,'kernel':'linear'}
