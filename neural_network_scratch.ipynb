```python
import numpy as np

# Sigmoid activation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Sigmoid derivative for backpropagation
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Mean Squared Error Loss
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Gradient of Mean Squared Error
def mse_derivative(y_true, y_pred):
    return y_pred - y_true

# Initialize weights and biases
def initialize_parameters(n_input, n_hidden, n_output):
    W1 = np.random.randn(n_hidden, n_input) * 0.01
    b1 = np.zeros((n_hidden, 1))
    W2 = np.random.randn(n_output, n_hidden) * 0.01
    b2 = np.zeros((n_output, 1))
    
    return W1, b1, W2, b2

# Forward propagation
def forward_propagation(X, W1, b1, W2, b2):
    Z1 = np.dot(W1, X) + b1
    A1 = sigmoid(Z1)
    Z2 = np.dot(W2, A1) + b2
    A2 = sigmoid(Z2)
    
    return Z1, A1, Z2, A2

# Backward propagation
def backward_propagation(X, Y, Z1, A1, Z2, A2, W2):
    m = X.shape[1]
    
    dZ2 = A2 - Y
    dW2 = np.dot(dZ2, A1.T) / m
    db2 = np.sum(dZ2, axis=1, keepdims=True) / m
    
    dA1 = np.dot(W2.T, dZ2)
    dZ1 = dA1 * sigmoid_derivative(Z1)
    dW1 = np.dot(dZ1, X.T) / m
    db1 = np.sum(dZ1, axis=1, keepdims=True) / m
    
    return dW1, db1, dW2, db2

# Update parameters
def update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate):
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    
    return W1, b1, W2, b2

# Train the neural network
def train(X, Y, n_hidden, epochs, learning_rate):
    n_input = X.shape[0]
    n_output = Y.shape[0]
    
    W1, b1, W2, b2 = initialize_parameters(n_input, n_hidden, n_output)
    
    for i in range(1,epochs):
        # Forward propagation
        Z1, A1, Z2, A2 = forward_propagation(X, W1, b1, W2, b2)
        
        # Compute loss
        loss = mean_squared_error(Y, A2)
        if i % 100 == 0:
            print(f"Epoch {i}, Loss: {loss}")
        
        # Backward propagation
        dW1, db1, dW2, db2 = backward_propagation(X, Y, Z1, A1, Z2, A2, W2)
        
        # Update weights and biases
        W1, b1, W2, b2 = update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate)
    
    return W1, b1, W2, b2

# Prediction function
def predict(X, W1, b1, W2, b2):
    _, _, _, A2 = forward_propagation(X, W1, b1, W2, b2)
    return np.argmax(A2, axis=0)

# Example Usage
if __name__ == "__main__":
    np.random.seed(1)
    
    # Example data (X: features, Y: labels)
    X = np.random.randn(2, 500)  # 500 samples, 2 features
    Y = (np.random.randn(1, 500) > 0).astype(int)  # 0 or 1 labels
    
    # Train the neural network
    W1, b1, W2, b2 = train(X, Y, n_hidden=4, epochs=1000, learning_rate=0.01)
    
    # Predict on new data
    predictions = predict(X, W1, b1, W2, b2)
    print(f"Predictions: {predictions}")
```
